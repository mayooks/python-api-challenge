```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

#import pprint
from pprint import pprint

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# Output File (CSV)
output_data_file = "output_data/cities.csv"
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="

#checking tthe url
pprint(requests.get(query_url + 'Harare').json())
```

```python
response = requests.get(query_url + city)
```

```python
towns = []
temps = []
temp_max = []
lats = []
lngs = []
humidities = []
cloudiness_s =[]
wind_speeds = []
countries = []
timestamps = []

for city in cities:
    
    
    
    response = requests.get(query_url + city)
    
    try:
        data = response.json()
        
        place = data['name']
        temp = data["main"]["temp"]
        maximum_temp = data["main"]["temp_max"]
        lat = data["coord"]["lat"]
        lang = data["coord"]["lon"]
        humidity = data["main"]['humidity']
        cloudiness =data["clouds"]["all"]
        wind_speed = data["wind"]["speed"]
        country = data["sys"]["country"]
        timestamp = data["dt"]
        
        
        
        print(" ")
        print(f"The city information recieved is {city}")
        print(f"The temperature for {city} is {temp}")
        print(f"The maximum temperature for {city} is {maximum_temp}")
        print(f"The latitude for {city} is {lat}")
        print(f"The longtitude for {city} is {lang}")
        print(f"The humidity for {city} is {humidity}")
        print(f"The wind speed for {city} is {wind_speed}")
        print(f"The country for {city} is {country}")
        print(f"The time for {city} is {timestamp}")
        print(" ")
        
        towns.append(place)
        temps.append(temp)
        #temp_max.append(maximum_temp)
        lats.append(lat)
        lngs.append(lang)
        humidities.append(humidity)
        cloudiness_s.append(cloudiness)
        wind_speeds.append(wind_speed)
        countries.append(country)
        timestamps.append(timestamp)
        
       
        
    except:
        print(f"\nFAILED: Got a response of {response.status_code} for city: {city}\n")
        
    
    time.sleep(2)


```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#putting together all the info into one dictionary
weather_dict = {'city':towns, 
                'temp':temps, 
                'lat':lats, 
                'lng':lngs, 
                'humidity':humidities,
                'cloudiness':cloudiness_s,
                'wind_speed':wind_speeds,
                'country':countries,
                'date' : timestamps
                }
```

```python
weather_data = pd.DataFrame(weather_dict)
```

## Linear Regression

```python
# creating dataframes for cities that are in the nothern and southern hemispheres 
# based on the latitue
nothern_hemisphere_df = weather_data.query("lat>0")
southern_hemisphere_df = weather_data.query("lat<0")
```

```python
nothern_hemisphere_df.head()
```

```python
southern_hemisphere_df.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
plt.figure(figsize=(12,8))
plt.scatter(data=nothern_hemisphere_df, x="lat", y="temp", 
            c = nothern_hemisphere_df['humidity'] );
sns.despine;


## defining x and y values for the linear regression model

x_values = nothern_hemisphere_df['lat'].astype('float')
y_values = nothern_hemisphere_df['temp'].astype('float')

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

#label plot
plt.title("Temperature in cities in the northern hemisphere vs latitude", 
          fontsize = 16, fontweight = 'bold')

plt.ylabel("Temperature (Celsius)", fontsize = 14, rotation = 90, 
           fontweight = 'bold', color = 'gray' )

plt.xlabel("latitude", fontsize = 14, fontweight = 'bold', color = 'gray')
#plt.legend(frameon = False, fontsize = 14)

# Plot regression line
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")

# inserting colorbar

cbar = plt.colorbar()
cbar.ax.set_ylabel('humidity', fontsize = 14, fontweight = 'bold', color = 'gray',
                   rotation=270)

plt.savefig("Temperature in cities in the nothern hemisphere vs latitue")

plt.show()

# Print r square value
print(f"R squared: {rvalue**2}")
```
